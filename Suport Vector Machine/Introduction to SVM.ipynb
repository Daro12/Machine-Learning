# Introduction to Support Vector Machine

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, [2, 3]].values # dependent variables
y = dataset.iloc[:, 4].values # independent variable
```

```python
X.shape
```

```python
X
```

```python
y.shape
```


## Splitting the dataset into the Training set and Test set

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```


## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```


## Fitting SVM to the Training set

```python
# start with the most simple kernel "Linear"
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```


## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

How can we tell if the model is good or not?


## Making the Confusion Matrix

A confusion matrix is a table that is often used to describe the performance of a classification model (or "classifier") on a set of test data for which the true values are known. Example:

<img src="confusion_matrix_simple.png" width="400" height="400">

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
```

## Visualising the Training set results

```python
from matplotlib.colors import ListedColormap  # help use to colorize the data point
X_set, y_set = X_train, y_train# create the local variable for plot

# take the min value of age/salary minus 1 so that the the axis lie below the min a little. 
# take the max value of age/salary plus 1 so that the the axis lie above the max a little. 
# the resolution is 0.01
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))


# for visualization of the classifier
# the contourf function is used to make the contour between class
# if the pixel belongs to the class 0 it colorizes black, 1 it colorizes white
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('black', 'white')))
```
